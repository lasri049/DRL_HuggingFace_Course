```python

!pip install gym[box2d]
!pip install stable-baselines3[extra]
!pip install huggingface_sb3
!pip install pyglet
from huggingface_sb3 import load_from_hub, push_to_hub, package_to_hub

from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env
```

```python
import gym
env = gym.make("LunarLander-v2")
observation = env.reset()

for _ in range(20):
  action = env.action_space.sample()
  print(f'Action taken: {action}')

  observation, reward, done, info = env.step(action)

if done:
  observation = env.reset()
  print('Environment is reset')
```

```python
print("_____OBSERVATION SPACE_____ \n")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample())
```

```python
env = make_vec_env('LunarLander-v2', n_envs = 16)
```

```python
model = PPO(
    policy = 'MlpPolicy',
    env = env,
    n_steps = 1024,
    batch_size = 64,
    n_epochs = 4,
    gamma = 0.999,
    gae_lambda = 0.98,
    ent_coef = 0.01,
    verbose=1)
```

```python
model.learn(total_timesteps=500000)
model.save('model\PPOPolicy')
```

```python
eval_env = gym.make('LunarLander-v2')
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic = True)
print(f'mean reward = {mean_reward:.2f}+/- {std_reward}')
```

```python
from huggingface_sb3 import package_to_hub

env_id = 'LunarLander-v2'
env = make_vec_env(env_id, n_envs=10)

eval_env = make_vec_env(env_id, n_envs=10)

model = PPO('MlpPolicy', env, n_steps=2000, batch_size=50, n_epochs=5, 
            gamma=0.99, verbose=1)
model.learn(total_timesteps=1000000)
package_to_hub(model=model,
               model_name = 'ppo-LunarLander-v2',
               model_architecture='PPO',
               env_id=env_id,
               eval_env=eval_env,
               repo_id='spacestar1705/ppo-LunaLander-v2',
               commit_message='new commit')
```
